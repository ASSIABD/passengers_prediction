```python
import pandas as pd
import numpy as np
import lightgbm as lgb
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

# Chargement du dataset

```python
df = pd.read_pickle("C:/Projets/Projets/ProjetML/clean_data/rider_ship_clean.pkl")
df.head()

```

```python
print(df.isna().sum())

# V√©rification des types de donn√©es
df.info()
```

# Suppression des colonnes inutiles

```python
df = df.drop(columns=["Date", "Number of Disembarking Passengers"])
df.head()
```

# Encodage des colonnes cat√©gorielles

```python

df["Day Week"] = df["Day Week"].astype("category")
df["Schedule Type"] = df["Schedule Type"].astype("category")

df.info()
```

# D√©finition des features et de la cible

```python
X = df[[
    "Annee", "Mois", "Jour",
    "Day Week",
    "Week Index", "Index Day Week",
    "Time Slot",
    "Schedule Type",
    "Number of Stops"
]]

y = df["Number of Boarding Passengers"]
```

# D√©coupage Train/Test

```python

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, shuffle=True, random_state=42
)
# Taille des donn√©es d'entrainement et de tes
X_train.shape, X_test.shape

```

# Cr√©ation et entra√Ænement du mod√®le LightGBM

```python

model_initial = lgb.LGBMRegressor(
    boosting_type="gbdt",
    n_estimators=500,
    learning_rate=0.05,
    max_depth=-1,
    num_leaves=31
)

model_initial.fit(X_train, y_train)
```

# Pr√©diction et √©valuation

```python
y_pred = model_initial.predict(X_test)

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print("RMSE :", rmse)
print("R2 Score :", r2)
```

```python
# V√©rification est ce que cette valeur de RMSE est acceptable ou non 
print("Mean Boarding Passengers :", y.mean())
print("Min Boarding Passengers :", y.min())
print("Max Boarding Passengers :", y.max())

rmse = 1647.5688633666484  # ta valeur RMSE
mean_target = y.mean()

pourcentage_erreur = (rmse / mean_target) * 100
print("Erreur moyenne (%) :", pourcentage_erreur)
```

```python
import numpy as np

def summarize_error(y_test, y_pred):
    """
    Calcule automatiquement les statistiques principales des erreurs
    entre valeurs r√©elles et valeurs pr√©dites.
    
    Retourne un dictionnaire avec :
    - moyenne, m√©diane, min, max
    - √©cart-type
    - RMSE
    - erreur relative moyenne (%)
    - nombre d'erreurs > 3*std (outliers)
    """
    errors = y_test - y_pred
    
    mean_error = np.mean(errors)
    median_error = np.median(errors)
    std_error = np.std(errors)
    min_error = np.min(errors)
    max_error = np.max(errors)
    rmse = np.sqrt(np.mean(errors**2))
    mean_target = np.mean(y_test)
    error_percent = (rmse / mean_target) * 100
    
    # Outliers : erreurs sup√©rieures √† 3*std
    threshold = 3 * std_error
    n_outliers = np.sum(np.abs(errors) > threshold)
    
    summary = {
        "Mean Error": mean_error,
        "Median Error": median_error,
        "Std Error": std_error,
        "Min Error": min_error,
        "Max Error": max_error,
        "RMSE": rmse,
        "Mean Target": mean_target,
        "Relative Error (%)": error_percent,
        "Outliers (>3œÉ)": n_outliers
    }
    
    return summary
```

```python
summarize_error(y_test, y_pred)
```

# V√©rification du sous et sur-apprentissage

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import learning_curve
from lightgbm import LGBMRegressor

# Remplacer les espaces dans les noms de colonnes pour √©viter les warnings
X.columns = [c.replace(' ', '_') for c in X.columns]

train_sizes, train_scores, test_scores = learning_curve(
    model_initial,
    X, y,
    train_sizes=np.linspace(0.1, 1.0, 10),
    cv=5,
    scoring='neg_root_mean_squared_error',  # RMSE n√©gatif
    shuffle=True,
    random_state=42
)

# Convertir en RMSE positif
train_scores = -train_scores
test_scores  = -test_scores

# Moyenne et √©cart type
train_scores_mean = np.mean(train_scores, axis=1)
train_scores_std  = np.std(train_scores, axis=1)
test_scores_mean  = np.mean(test_scores, axis=1)
test_scores_std   = np.std(test_scores, axis=1)





# Trac√© des courbes
plt.figure(figsize=(6,4))
plt.plot(train_sizes, train_scores_mean, 'o-', label='Train RMSE')
plt.plot(train_sizes, test_scores_mean, 'o-', label='Validation RMSE')
plt.fill_between(train_sizes,
                 train_scores_mean - train_scores_std,
                 train_scores_mean + train_scores_std,
                 alpha=0.1)
plt.fill_between(train_sizes,
                 test_scores_mean - test_scores_std,
                 test_scores_mean + test_scores_std,
                 alpha=0.1)

plt.xlabel("Taille de l'√©chantillon d'entra√Ænement")
plt.ylabel("RMSE")
plt.title("Courbes d'apprentissage LightGBM")
plt.legend(loc="best")
plt.show()


```

# Optimisation du mod√®le

```python
model_reg = lgb.LGBMRegressor(
    boosting_type="gbdt",
    n_estimators=500,
    learning_rate=0.05,
    max_depth=6,
    num_leaves=31,
    min_child_samples=30,
    reg_alpha=0.1,
    reg_lambda=0.1
)

model_reg.fit(X_train, y_train)
y_pred_reg = model_reg.predict(X_test)
rmse_reg = np.sqrt(mean_squared_error(y_test, y_pred_reg))
r2_reg = r2_score(y_test, y_pred_reg)
print("RMSE apr√®s r√©gularisation :", rmse_reg)
print("R2 Score apr√®s r√©gularisation :", r2_reg)
```

```python
# Cell: Tester l'overfitting
from sklearn.metrics import mean_squared_error, r2_score

# Pr√©dictions sur train et test
y_pred_train = model_reg.predict(X_train)
y_pred_test = model_reg.predict(X_test)

# Calcul des m√©triques
rmse_train = np.sqrt(mean_squared_error(y_train, y_pred_train))
rmse_test = np.sqrt(mean_squared_error(y_test, y_pred_test))
r2_train = r2_score(y_train, y_pred_train)
r2_test = r2_score(y_test, y_pred_test)

# Affichage
print("RMSE Train :", rmse_train)
print("RMSE Test  :", rmse_test)
print("Diff√©rence RMSE (Test-Train) :", rmse_test - rmse_train)
print("R¬≤ Train :", r2_train)
print("R¬≤ Test  :", r2_test)
print("Diff√©rence R¬≤ (Train-Test) :", r2_train - r2_test)


```

## Analyse des performances du mod√®le r√©gularis√©
- Les courbes d'apprentissage


```python

# Diviser le train initial en X_train_small + X_val pour validation
X_tr, X_val, y_tr, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

# Tailles d‚Äô√©chantillons
train_sizes = np.linspace(0.1, 1.0, 10)

train_scores = []
val_scores = []

for frac in train_sizes:
    n_samples = int(frac * X_tr.shape[0])
    X_sub = X_tr[:n_samples]
    y_sub = y_tr[:n_samples]
    
    model_temp = lgb.LGBMRegressor(
        boosting_type="gbdt",
        n_estimators=500,
        learning_rate=0.05,
        max_depth=6,
        num_leaves=31,
        min_child_samples=30,
        reg_alpha=0.1,
        reg_lambda=0.1
    )
    
    model_temp.fit(X_sub, y_sub)
    
    y_train_pred = model_temp.predict(X_sub)
    y_val_pred = model_temp.predict(X_val)
    
    train_scores.append(np.sqrt(mean_squared_error(y_sub, y_train_pred)))
    val_scores.append(np.sqrt(mean_squared_error(y_val, y_val_pred)))

# Plot
plt.figure(figsize=(6,4))
plt.plot(train_sizes * 100, train_scores, 'o-', color='blue', label='RMSE Train')
plt.plot(train_sizes * 100, val_scores, 'o-', color='red', label='RMSE Validation')
plt.xlabel("Taille du jeu de train (%)")
plt.ylabel("RMSE")
plt.title("Courbes d'apprentissage - LightGBM r√©gularis√©")
plt.legend()
plt.show()
```

```python
def extraire_valeurs_courbes(train_sizes, train_scores, val_scores):
    """
    Retourne les valeurs des courbes d'apprentissage.

    Param√®tres :
    - train_sizes : liste ou array des tailles d'entrainement (en nombre d'exemples ou fraction)
    - train_scores : RMSE ou m√©trique train correspondante
    - val_scores : RMSE ou m√©trique validation correspondante

    Retour :
    - dict contenant :
        - 'train_sizes' : tailles utilis√©es pour le train
        - 'train_scores' : valeurs RMSE train
        - 'val_scores' : valeurs RMSE validation
    """
    return {
        'train_sizes': np.array(train_sizes),
        'train_scores': np.array(train_scores),
        'val_scores': np.array(val_scores)
    }


valeurs_courbes = extraire_valeurs_courbes(train_sizes*len(X_tr), train_scores, val_scores)
print(valeurs_courbes)
```

## Analyse de la courbe d'apprentissage 

1Ô∏è‚É£ Tendances observ√©es

Le mod√®le apprend progressivement, l‚Äôerreur sur le train augmente un peu car les tailles plus grandes r√©duisent l‚Äôeffet d‚Äôoverfitting sur un petit √©chantillon.

<strong>RMSE validation :</strong>

Cela montre que plus de donn√©es aident le mod√®le √† mieux g√©n√©raliser.

2Ô∏è‚É£ Comparaison train vs validation

√âcart final RMSE (Validation - Train) = 1781 ‚àí 1665 ‚âà 116

Tr√®s faible par rapport √† la valeur absolue (~1780), donc le mod√®le est stable et ne sur-apprend pas.

√âcart initial RMSE (Validation - Train) = 2498 ‚àí 1539 ‚âà 959

Au d√©but, avec peu de donn√©es, le mod√®le sur-apprenait un peu ‚Üí normal.

3Ô∏è‚É£ Interpr√©tation g√©n√©rale

Tendance RMSE train : l√©g√®re augmentation ‚Üí normal avec plus de donn√©es, r√©duit le surapprentissage.

Tendance RMSE validation : nette diminution ‚Üí montre que le mod√®le gagne en g√©n√©ralisation avec plus de donn√©es.

Diff√©rence finale faible ‚Üí le mod√®le r√©gularis√© est stable, fiable, et g√©n√©ralise bien sur de nouvelles donn√©es.

Conclusion : pas de surapprentissage, pas de sous-apprentissage significatif, mod√®le performant.
